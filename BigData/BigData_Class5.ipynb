```python
import pandas as pd
```

```python
# Load the Excel file into a DataFrame
df = pd.read_csv('/content/agg_trans.csv')
```

```python
# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
# Calculate the count of unique regions
unique_region_count = df['Region'].nunique()
```

```python
# Get the list of unique regions
unique_regions = df['Region'].unique()
```

```python
print(f"Count of unique Regions: {unique_region_count}")
print(f"Unique Regions: {list(unique_regions)}")
```

```python
# Filter the DataFrame for the 'North-Eastern Region'
northeast_df = df[df['Region'] == 'North-Eastern Region']
```

```python
# Calculate the count of unique states in the 'North-Eastern Region'
unique_states_northeast_count = northeast_df['State'].nunique()
```

```python
# Get the list of unique states in the 'North-Eastern Region'
unique_states_northeast = northeast_df['State'].unique()
```

```python
print(f"Count of unique States in North-Eastern Region: {unique_states_northeast_count}")
print(f"Unique States in North-Eastern Region: {list(unique_states_northeast)}")
```

```python
# Get the unique transaction types from the 'Transaction_type' column
unique_transaction_types = df['Transaction_type'].unique()
```

```python
print(f"Different Transaction Types: {list(unique_transaction_types)}")
```

```python
# Calculate the total transaction amount for each year
year_transaction_summary = df.groupby('Year')['Transaction_amount'].sum()
```

```python
# Find the year with the highest transaction amount
year_with_highest_transaction = year_transaction_summary.idxmax()
highest_transaction_amount = year_transaction_summary.max()
```

```python
print(f"The year with the highest transaction amount is {year_with_highest_transaction} with a total of {highest_transaction_amount:.2f}.")
```

```python
# Calculate the total transaction count for each transaction type
transaction_type_summary = df.groupby('Transaction_type')['Transaction_count'].sum()
```

```python
# Find the transaction type with the lowest transaction count
transaction_type_lowest_count = transaction_type_summary.idxmin()
lowest_transaction_count = transaction_type_summary.min()
```

```python
print(f"The transaction type with the lowest transaction count is '{transaction_type_lowest_count}' with a total of {lowest_transaction_count:.0f} transactions.")
```
